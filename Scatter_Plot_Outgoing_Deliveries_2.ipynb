```python
# Install Pandas Library:

import pandas as pd

# Google Drive Library:

from google.colab import drive
drive.mount('/content/drive')

# Actual Path In Google Drive ie: /content/drive/....(File Path) :

file_path = '/content/drive/My Drive/Data_Anayltics/Simple Linear Regression_Outgoing Deliveries.xlsx'

# Read the excel
df = pd.read_excel(file_path)

# Visualization Plot Chart - Matplotlib
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.scatter(df['Distance (km)'], df['Delivery_Time_Hr'], marker='o', color='royalblue')

plt.title('Delivery_Time_Hr vs Distance (km)')
plt.xlabel('Distance (km)')
plt.ylabel('Delivery_Time_Hr')
plt.grid()
plt.tight_layout()
plt.show()
```
